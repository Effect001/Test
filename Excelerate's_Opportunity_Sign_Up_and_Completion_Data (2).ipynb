# Statement of Objective

The primary aim of this project is to analyze and derive actionable insights from Excelerate's Opportunity Sign Up and Completion Data. This analysis will focus on understanding user engagement patterns, demographic distributions, and the effectiveness of various opportunities offered by Excelerate. The insights garnered will help Excelerate refine their opportunity offerings, tailor their engagement strategies to meet user needs, and ultimately enhance the overall user experience on their platform.


# Our Objectives

1. To ensure the dataset is accurate, consistent, and free from anomalies by conducting thorough data cleaning and preprocessing.

2. To explore the demographic composition of participants, such as gender distribution, student status, and geographic location. This will help Excelerate understand who is engaging with their platform and identify any underrepresented groups.

3. To analyze the distribution and engagement levels across different opportunity categories (e.g., internships, courses, events).

3. To assess the distribution of rewards and skill points earned by participants. Understanding these distributions will help Excelerate evaluate the incentive structures of their opportunities and explore ways to optimize them.

4. To examine the uniqueness of Profile IDs and identify patterns of repeated engagement. This will shed light on user retention and repeat participation, offering insights into user loyalty and long-term engagement.

5. To analyze the status distribution of opportunities (e.g., applied, in progress, completed) and understand participant progression through various stages.

6. To create clear and insightful visualizations that effectively communicate key findings.

# Keywords

- Data Cleaning: The process of detecting and correcting or removing corrupt or inaccurate records from a dataset.

- Demographic Analysis: The study of populations based on factors such as age, race, gender, economic status, and level of education.

- Engagement Patterns: Trends and behaviors exhibited by users in their interactions with opportunities or platforms.

- Opportunity Categories: Various classifications of opportunities provided by Excelerate, including internships, courses, events, and more.

- Reward Amount: The monetary or non-monetary incentives offered to participants upon completing certain opportunities.

- Skill Points: Points earned by participants that reflect their engagement level or skill acquisition through opportunities.

- Profile ID: A unique identifier assigned to each participant in the dataset.

- Opportunity Status: The current phase or progress stage of an opportunity, such as applied, in progress, or completed.

- Visualization: The representation of data through graphical elements such as charts, graphs, and plots to communicate insights effectively.

- Dashboard: An interactive tool that displays key metrics and visualizations to provide an overview of data insights and trends.

```python
# Calculate basic statistics for Reward Amount
reward_stats = data['Reward Amount'].describe()

# Calculate basic statistics for Skill Points Earned
skill_stats = data['Skill Points Earned'].describe()

# Print out the results
print("Basic Statistics for Reward Amount:")
print(reward_stats)

print("\nBasic Statistics for Skill Points Earned:")
print(skill_stats)
```

```python
# Calculate mean, mode, and median for 'Reward Amount'
column_mean_reward = data['Reward Amount'].mean()
column_mode_reward = data['Reward Amount'].mode()
column_median_reward = data['Reward Amount'].median()

# Calculate mean, mode, and median for 'Skill Points Earned'
column_mean_skill = data['Skill Points Earned'].mean()
column_mode_skill = data['Skill Points Earned'].mode()
column_median_skill = data['Skill Points Earned'].median()

# Printing
print("Reward Amount:")
print(f"Mean: {column_mean_reward}")
print(f"Mode: {column_mode_reward}")
print(f"Median: {column_median_reward}")

print("\nSkill Points Earned:")
print(f"Mean: {column_mean_skill}")
print(f"Mode: {column_mode_skill}")
print(f"Median: {column_median_skill}")
```

```python
# Replace missing values in 'Reward Amount' with median
data['Reward Amount'] = data['Reward Amount'].fillna(data['Reward Amount'].median())

# Replace missing values in 'Skill Points Earned' with median
data['Skill Points Earned'] = data['Skill Points Earned'].fillna(data['Skill Points Earned'].median())
```

```python
#Confirmation of data for any missing value on Reward Amount and Skill Points Earned after replacing initial missed values
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
# Handling missing values in 'Gender', 'City', 'State', 'Zip Code', 'Graduation Date(YYYY MM)', 'Current Student Status'
#Replacing missing values with unkown
data['Gender'].fillna('Unknown', inplace=True)
data['City'].fillna('Unknown', inplace=True)
data['State'].fillna('Unknown', inplace=True)
data['Zip Code'].fillna('Unknown', inplace=True)
data['Graduation Date(YYYY MM)'].fillna('Unknown', inplace=True)
data['Current Student Status'].fillna('Unknown', inplace=True)
data['Current/Intended Major'].fillna('Unknown', inplace=True)

data['Badge Id'].fillna('Unknown', inplace=True)
data['Badge Name'].fillna('Unknown', inplace=True)
data['Skills Earned'].fillna('Unknown', inplace=True)

data['Opportunity Start Date'].fillna(method='ffill', inplace=True)
```

```python
#Checking for missing values after replacement
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
#To check for our data shape, Unique and duplicate 
print("The shape of our data: ")
print(data.shape[0])

print("\nNumber of unique Profile IDs: ")
print(data['Profile Id'].nunique())

print("\nNumber of duplicated Profile IDs: ")
print(data.duplicated(subset=['Profile Id']).sum())
```


We can see there are 8,841 duplicated Profile IDs out of 20,322 total records in our dataset. This indicates that there are instances where the same Profile ID appears more than once in our data.

# Visualization

```python
#Visualization of Distribution of Opportunity Status

# Count occurrences of each status description
status_counts = data['Status Description'].value_counts()

# Display distribution summary
print("Opportunity Status Distribution Summary:")
print(status_counts)

# Plotting the distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=status_counts.index, y=status_counts.values, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Status Description')
plt.ylabel('Count')
plt.title('Distribution of Opportunity Statuses')
plt.tight_layout()
plt.show()
```

-- From the chart, it shows that Team allocation has the highest distribution of oppoturninty status with Rewards Award next

```python
#To show statistics of opportunity Name Frequency Distribution using Histogram
plt.figure(figsize= (12, 6))
sns.histplot(data['Opportunity Name'], bins=20, kde=True)
plt.title('Opportunity Name Frequency Distribution')
plt.xlabel('Opportunity Name')
plt.ylabel('frequency')
plt.xticks(rotation=90)
plt.show()
```

This histogram shows the frequency distribution of different Opportunity Names in the dataset.

From the graph, it is evident that "Data Visualization" is the most frequent Opportunity Name, indicating it appears most often compared to other opportunities.

```python
#To show statistics of opportunity Category Frequency Distribution using Histogram
plt.figure(figsize= (12, 6))
sns.histplot(data['Opportunity Category'], bins=20, kde=True)
plt.title('Opportunity Category Frequency Distribution')
plt.xlabel('Opportunity Category')
plt.ylabel('frequency')
plt.xticks(rotation=90)
plt.show()
```

This histogram illustrates the frequency distribution of Opportunity Categories across the dataset.

The graph highlights that "Internship" is the most common Opportunity Category based on frequency, suggesting it occurs more frequently than other categories such as Event, Course, Competition, and Engagement.

```python
#To check distribution of opportunity categories across various opportunity name

plt.figure(figsize=(12, 6))
hue_order = ['Event', 'Course', 'Competition', 'Internship', 'Engagement']
sns.countplot(data=data, x='Opportunity Name', hue='Opportunity Category', hue_order = hue_order)
plt.title('Opportunity Category Distribution by Opportunity Name')
plt.xlabel('Opportunity Name')
plt.ylabel('Opportunity Category Count')
plt.legend(['Event', 'Course', 'Competition', 'Internship', 'Engagement'])
plt.legend(title='Opportunity Category', labels=hue_order)
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

The above count plot visualizes the distribution of Opportunity Categories across various Opportunity Names.

- The graph displays how different Opportunity Categories (Event, Course, Competition, Internship, Engagement) are distributed across specific Opportunity Names.
- It is notable that "Internship" (colored red) is prominently associated with various Opportunity Names such as Data Visualization, Digital Marketing, Health Care Management, Innovation and Entrepreneurship, and Project Management.
- This indicates that Internship opportunities span across a diverse set of fields or subjects within the dataset.

```python
#To get the current student status distribution


# Calculate the frequency of each category
status_counts = data['Current Student Status'].value_counts().sort_values(ascending=False)

# Plotting the bar plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=status_counts.index, y=status_counts.values)
plt.title('Current Student Status Frequency Distribution')
plt.xlabel('Current Student Status')
plt.ylabel('Frequency')
plt.xticks(rotation=90)

# Annotate the bars with the counts
for i in ax.containers:
    ax.bar_label(i, label_type='edge')

#To show the plot
plt.show()
```

- This is followed by undergraduate students, numbering 7009.
- Other statuses such as 'Unknown', have significantly lower frequencies.
